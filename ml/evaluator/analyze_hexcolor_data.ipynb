```python
import pickle
from pathlib import Path
import os

# Define path to hexcolor_vf directory
# Assuming this notebook is in ml/evaluator/
HEXCOLOR_DIR = Path('../../hexcolor_vf')

print(f"Looking for .pkl files in: {HEXCOLOR_DIR.resolve()}")
```

```python
def count_items_in_pkl(file_path):
    try:
        with open(file_path, 'rb') as f:
            data = pickle.load(f)
        
        if isinstance(data, list):
            return len(data), type(data).__name__
        elif hasattr(data, '__len__'):
            return len(data), type(data).__name__
        else:
            return "N/A (Not a list/sized)", type(data).__name__
            
    except Exception as e:
        return f"Error: {str(e)}", "Error"

# Find and process all .pkl files
pkl_files = sorted(HEXCOLOR_DIR.glob('*.pkl'))

print(f"{'Filename':<30} | {'Count':<10} | {'Type':<15}")
print("-" * 60)

results = []
for pkl_file in pkl_files:
    count, data_type = count_items_in_pkl(pkl_file)
    print(f"{pkl_file.name:<30} | {str(count):<10} | {data_type:<15}")
    results.append({'file': pkl_file.name, 'count': count, 'type': data_type})
```
